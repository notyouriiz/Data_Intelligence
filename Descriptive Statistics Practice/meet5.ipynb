# **Latihan (Probabilitas)**

Asosiasi  Olah Raga Nasional melakukan survei terhadap orang-orang berusia 7 tahun atau lebih tua tentang partisipasi dalam kegiatan olahraga (Abstrak Statistik Amerika Serikat: 2002).
Jumlah penduduk pada kelompok umur ini dilaporkan sebanyak 248,5 juta, dengan 120,9 juta laki-laki dan 127,6 juta perempuan. Jumlah peserta untuk lima aktivitas olahraga teratas adalah \\

|Activity| Male |Female|
|----|----|-----|
Bicycle riding |22.2| 21.0
Camping |25.6 |24.3
Exercise walking| 28.7 |57.7
Exercising with equipment |20.4 |24.4
Swimming |26.4| 34.4

a. Untuk wanita yang dipilih secara acak, perkirakan probabilitas partisipasi di masing-masing 5 kegiatan olahraga? <br>
b. Untuk orang yang dipilih secara acak, berapa probabilitas orang tersebut berpartisipasi olahraga excercise walking?

```python
import pandas as pd

# Menerapkan dictionary untuk membuat dataframe
olahraga_data = pd.DataFrame(
    {
        "Activity": [
            "Bicycle Riding",
            "Camping",
            "Exercise Walking",
            "Exercising with equipment",
            "Swimming",
        ],
        "Male": [22.2, 25.6, 28.7, 20.4, 26.4],
        "Female": [21.0, 24.3, 57.7, 24.4, 34.4],
    }
)

# Display DataFrame
olahraga_data

# totaling male and female
olahraga_data['Total'] = olahraga_data['Male'].sum() + olahraga_data['Female'].sum()


olahraga_data
```

```python
# probabilities
#olahraga_data['probability_male'] = olahraga_data['Male'] / olahraga_data['Total Male']
olahraga_data['probability_female'] = olahraga_data['Female'] / olahraga_data['Total']
olahraga_data
```

```python
# Hitung probabilitas olahraga Exercise Walking
total_male_exercise_walking = 28.7
total_female_exercise_walking = 57.7

total_male = olahraga_data['Male'].sum()
total_female = olahraga_data['Female'].sum()

probabilitas_exercise_walking = (total_male_exercise_walking + total_female_exercise_walking) / (total_male + total_female)

print("Probabilitas Exercise Walking:", probabilitas_exercise_walking)
```

# Statistika Deskriptif di bisnis

Berdasarkan data transaksi dari coffee shop. Carilah persebaran umur berdasarkan mean, dan varian dari customer tersebut. Serta dapatkan modus usia dari customers.


## Persebaran umur berdasarkan mean


```python
import pandas as pd
import statistics
import datetime as dt

# calculates customer age by year only
current_year = 2024
customer["age"] = current_year - customer["birth_year"]


customer["birthdate"] = pd.to_datetime(
    customer["birthdate"]
)  # Convert 'birthdate' to datetime
current_date = dt.datetime.now()  # Calculate customer age in years
customer["usia"] = (
    current_date - customer["birthdate"]
).dt.days // 365  # calculate current date - total of cust birthdate day divide by days in a year

# mean of age
mean_age = customer["age"].mean()

# age variance
var_age = customer["age"].var()

# count
mode_customer = statistics.mode(customer["age"])

# print results
print("Age by Years: ", customer["age"])
print("mean age: ", mean_age)
print("var age: ", var_age)
print("mode age: ", mode_customer)
print("Age by Datetime: ", customer["usia"])
```
