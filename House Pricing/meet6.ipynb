## Importing File

```python
import pandas as pd
import seaborn as seaborn

house_data = pd.read_csv('house_pricing.csv')
house_data

house_data.describe().style.background_gradient(cmap='BuPu')
```

## Pertanyaan 1
Mencari rata-rata dan standar deviasi dari luas area dan luas garasi yang ditawarkan.

```python
# mean
# Luas Area rumah
area_mean = house_data["GrLivArea"].mean()
print ('Rerata Luas Area: ', area_mean)

#luas area garasi
garage_mean = house_data["GarageArea"].mean()
print ('Rerata Luas Area Garasi: ', garage_mean)

# std dev Area rumah
area_std = house_data["GrLivArea"].std()
print ('Rerata Luas Area: ', area_std)

#std dev area garasi
garage_std = house_data["GarageArea"].std()
print ('Rerata Luas Area Garasi: ', garage_std)
```

## Pertanyaan 2
Mencari rata-rata luas area dan garasi berdasarkan neighborhood dan street

```python
# Convert LotArea and GarageArea columns to numeric
house_data["GrLivArea"] = pd.to_numeric(house_data["GrLivArea"], errors="coerce")
house_data["GarageArea"] = pd.to_numeric(house_data["GarageArea"], errors="coerce")

# Group by Neighborhood and Street
group_neighborhood_streer_data = house_data.groupby(["Neighborhood", "Street"])

# Calculate average GrLivArea and GarageArea
avg_areas = group_neighborhood_streer_data.agg(
    {"GrLivArea": "mean", "GarageArea": "mean"}
).reset_index()

# using reset inex is basically for reseting the level to make it easier to code when we need to analyze or manipulate the data in the next time
# when we not using index it would be have multiple level and we need to specify it for example 'GarageArea', 'Mean' | while if we using reset index we can just say "GarageArea"

# Display the result
print(avg_areas)
```

## Pertanyaan 3
Mencari kisaran luas area dan garasi untuk jalan dengan tipe pave

```python
# Filter data for street type "Pave"
pave_data = house_data[house_data['Street'] == 'Pave']

# Calculate the range of LotArea and GarageArea
lotarea_range = (pave_data['LotArea'].min(), pave_data['LotArea'].max())
garagearea_range = (pave_data['GarageArea'].min(), pave_data['GarageArea'].max())

# Display the result
print("Range of LotArea for Paved Streets:", lotarea_range)
print("Range of GarageArea for Paved Streets:", garagearea_range)
```
