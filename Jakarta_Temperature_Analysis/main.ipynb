# STEP 3 - MEMBUAT VISUALISASI

# VISUALISASI PERTAMA

Perkembangan Suhu Rata-Rata di Jakarta pada Tahun 1995 - 2019

Mengelompokkan DataFrame jakarta_df berdasarkan tahun(Year) , menghitung rata-rata(mean) pada tiap kelompok tahunnya , dan menamai masing-masing baris dengan index yang dimulai dari 0

```python
meanTemperature_df = jakarta_df.groupby('Year').aggregate(
    {'AvgTemperature' : 'mean'}
).reset_index()
```

Mengambil data hanya pada tahun 1995 - 2019, karena data csv pada tahun 2020 tidak lengkap

```python
meanTemperature_df = meanTemperature_df.iloc[0:25:,[0,1]]
```

```python
meanTemperature_df
```

Mencari value suhu tertinggi 

```python
indeks_suhu_tertinggi = meanTemperature_df['AvgTemperature'].idxmax()

suhu_tertinggi = meanTemperature_df['AvgTemperature'].max()

tahun_tertinggi = meanTemperature_df['Year'].loc[indeks_suhu_tertinggi]

print(indeks_suhu_tertinggi)
print(suhu_tertinggi)
print(tahun_tertinggi)
```

Mencari value suhu terendah

```python
indeks_suhu_terendah = meanTemperature_df['AvgTemperature'].idxmin()

suhu_terendah = meanTemperature_df['AvgTemperature'].min()

tahun_terendah = meanTemperature_df['Year'].loc[indeks_suhu_terendah]

print(indeks_suhu_terendah)
print(suhu_terendah)
print(tahun_terendah)
```

```python
visualization1 = meanTemperature_df.plot(x='Year', xticks=range(1995, 2020, 4), kind='line', marker='o', legend=False)

visualization1.set_xlabel('Tahun')
visualization1.set_ylabel('Suhu rata-rata')

visualization1.set_title('Perkembangan Suhu Rata-Rata di Jakarta pada Tahun 1995 - 2019')
```

# VISUALISASI KEDUA

Perkembangan Suhu Rata-Rata di Jakarta Tiap Bulan pada Tahun 2002 (Terendah pada periode 1995 - 2019)

```python
suhu2002_df = jakarta_df[jakarta_df['Year'] == 2002].reset_index()

suhu2002_df = suhu2002_df[['Year', 'Month', 'AvgTemperature']]

suhu2002_df
```

```python
meanMonthIn2002 = suhu2002_df.groupby('Month').aggregate(
    {'AvgTemperature' : 'mean'}
).reset_index()

meanMonthIn2002
```

```python
visualization2 = meanMonthIn2002.plot(x='Month', kind='line', marker='o', legend=False, xticks=range(1, 13))

visualization2.set_xlabel('Bulan')
visualization2.set_ylabel('Suhu rata-rata')

visualization2.set_title('Perkembangan Suhu Rata-Rata di Jakarta selama 12 Bulan di Tahun 2002')

bulan_labels = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
visualization2.set_xticklabels(bulan_labels, rotation=45)
```

# VISUALISASI KETIGA

Perkembangan Suhu Rata-Rata di Jakarta Tiap Bulan pada Tahun 2011 (Tertinggi pada periode 1995 - 2019)

```python
suhu2011_df = jakarta_df[jakarta_df['Year'] == 2011].reset_index()

suhu2011_df = suhu2011_df[['Year', 'Month', 'AvgTemperature']]

suhu2011_df
```

```python
meanMonthIn2011 = suhu2011_df.groupby('Month').aggregate(
    {'AvgTemperature' : 'mean'}
).reset_index()

meanMonthIn2011
```

```python
visualization3 = meanMonthIn2011.plot(x='Month', kind='line', marker='o', legend=False,xticks=range(1, 13))

visualization3.set_xlabel('Bulan')
visualization3.set_ylabel('Suhu rata-rata')

visualization3.set_title('Perkembangan Suhu Rata-Rata di Jakarta selama 12 Bulan di Tahun 2011')

visualization3.set_xticklabels(bulan_labels, rotation=45)
```

# VISUALISASI KEEMPAT

Perkembangan Suhu Rata-Rata di Jakarta Pada Bulan September pada Tahun 2011 (Tertinggi pada tahun 2011)

```python
tempInSeptember2011_df = jakarta_df[(jakarta_df['Year'] == 2011) & (jakarta_df['Month'] == 9)].reset_index()

tempInSeptember2011_df = tempInSeptember2011_df[['Day','AvgTemperature']]

tempInSeptember2011_df
```

```python
visualization3 = tempInSeptember2011_df.plot(x='Day', kind='line', marker='o', legend=False, xticks=range(1, 31, 4))

visualization3.set_xlabel('Tanggal')
visualization3.set_ylabel('Suhu rata-rata')

visualization3.set_title('Perkembangan Suhu Rata-Rata di Jakarta pada Bulan September di Tahun 2011')
```
