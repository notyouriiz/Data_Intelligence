```python
import pandas as pd
from scipy import stats
import seaborn as sn
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
house_price = pd.read_csv('house_pricing.csv')
house_price.head().style.background_gradient(cmap='BuPu')
```

### No. 1
Uji hipotesis dengan asumsi bahwa luas area yang ditawarkan tidak kurang dari 9400

```python
# Ambil kolom LotArea dari DataFrame
lot_area = house_price['LotArea']

# Nilai yang diuji
uji_value = 9400

# Lakukan one-sample t-test
t_statistic, p_value = stats.ttest_1samp(lot_area, uji_value, alternative='greater')

# Hitung degree of freedom
degree_freedom = len(lot_area) - 1

# Tampilkan hasil
print("t-statistic:", t_statistic)
print("p-value:", p_value)
print("Degree of freedom:", degree_freedom)
```

### No. 2
pengaruh tipe jalan terhadap luas area dan harga rumah

```python
# Membagi data menjadi kelompok berdasarkan tipe jalan
grouped_data = house_price.groupby('Street')

# Melakukan uji ANOVA untuk luas area
model_area = ols('LotArea ~ Street', data=house_price).fit()
anova_area = sm.stats.anova_lm(model_area, typ=2)

# Melakukan uji ANOVA untuk harga rumah
model_price = ols('SalePrice ~ Street', data=house_price).fit()
anova_price = sm.stats.anova_lm(model_price, typ=2)

# Tampilkan hasil
print("ANOVA for LotArea:")
print(anova_area)
print("\nANOVA for SalePrice:")
print(anova_price)
```
