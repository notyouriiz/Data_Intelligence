```python
plt.figure(figsize=(20, 10))

# First subplot
plt.subplot(2, 2, 1)
plt.title('Gender')
ax1 = sns.countplot(x='Gender', palette='mako', data=sleep_data)
for p in ax1.patches:
    ax1.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Second subplot
plt.subplot(2, 2, 2)
plt.title('BMI Category')
ax2 = sns.countplot(x='BMI Category', palette='mako', data=sleep_data)
for p in ax2.patches:
    ax2.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Third subplot
plt.subplot(2, 2, 3)
plt.title('Sleep Disorder')
ax3 = sns.countplot(x='Sleep Disorder', palette='mako', data=sleep_data)
for p in ax3.patches:
    ax3.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
sleep_data['Sleep Disorder'] = sleep_data['Sleep Disorder'].fillna('None')
```

```python
# Calculate the percentage values for each category
gender_counts = sleep_data['Gender'].value_counts()

plt.figure(figsize=(30, 20), facecolor='#0C134F')

# Define the 'mako' color palette
colors = sns.color_palette('mako')

# Define a bigger font size
font = {'size': 16}

# Set the font size for the entire plot
plt.rc('font', **font)

# Define text properties with white font color
text_props = {'color': 'white'}

# First subplot - Pie chart for Gender
plt.subplot(2, 2, 1)
plt.title('Gender', fontsize=30, color='white')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=colors, textprops=text_props)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(gender_counts.index, title="Gender", loc="best")

plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage values for each category
bmi_counts = sleep_data['BMI Category'].value_counts()

plt.figure(figsize=(30, 20), facecolor='#0C134F')

# Define the 'mako' color palette
colors = sns.color_palette('mako')

# Define a bigger font size
font = {'size': 16}

# Set the font size for the entire plot
plt.rc('font', **font)

# Define text properties with white font color
text_props = {'color': 'white'}

# Second subplot - Pie chart for BMI Category
plt.subplot(2, 2, 2)
plt.title('BMI Category', fontsize=20, color='white', y=1.08)  # Adjust y parameter
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', startangle=90, colors=colors, textprops=text_props)
plt.axis('equal')
plt.legend(bmi_counts.index, title="BMI Category", loc="best")

plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage values for each category
gender_counts = sleep_data['Gender'].value_counts()

plt.figure(figsize=(30, 20), facecolor='#0C134F')

# Define the 'mako' color palette
colors = sns.color_palette('mako')

# Define a bigger font size
font = {'size': 16}

# Set the font size for the entire plot
plt.rc('font', **font)

# Define text properties with white font color
text_props = {'color': 'white'}

sleep_data.groupby('Sleep Disorder')['Gender'].value_counts().plot.pie(
                                                                    autopct ='%1.1f%%',figsize=(15,7),
                                                                    colors=colors,
                                                                    textprops=text_props,
                                                                    )

plt.title('The relationship between (Sleep Disorder, Gender)', color='white', fontsize=20)
plt.axis('equal')
plt.show()
```
